```python
# Question 7: Find the factorial of 8

num = int(input("Enter a number: "))
factorial = 1
for i in range (1, num+1):
    factorial *= i
print (factorial)
```

```python
# Question 8: Calculate the average of numbers: 15, 23, 31, 42, 56

num = [15, 23, 31, 42, 56]
lenght = len(num)
sum = 0
for i in num:
    sum = sum+i
avg = sum / lenght
print(avg)
```

```python
# Question 10: Calculate the sum of first 20 odd numbers

# Number of terms
n = 20

# Generate the first 20 odd numbers and sum them
# range(start, stop, step)
# The 20th odd number is (2 * 20) - 1 = 39, so stop at 40

odd_numbers = range(1, 2 * n, 2)
total_sum = sum(odd_numbers)

print(f"The sum of the first {n} odd numbers is: {total_sum}")
```
