```python
# Exercise 1: Simple Age Check Write a program that checks if a person can get a driver's license (age >= 16).

age = int(input("Enter your age : "))

if age >= 40:
    print(f"Your age is {age}.")
    print("You can get a driver's license.")
    print("But a mandatory medical certificate (Form 1A) required for each renewal to verify vision and physical fitness. ")  
elif age >= 20:
    print(f"Your age is {age}.")
    print("You can get a driver's license")
    print("You are eligible to drive transport/commercial vehicles, such as taxis, buses, and heavy goods vehicles. ")
elif age >= 18:
    print(f"Your age is {age}.")
    print("You can get a driver's license")
    print("Ypu are eligible for Light Motor Vehicles (LMV), including cars and motorcycles with gears.") 
elif age >= 16:
    print(f"Your age is {age}.")
    print("You can get a driver's license")
    print("Your eligible for a gearless scooter/motorcycle with an engine capacity not exceeding 50cc.")
elif age <= 0:
    print(f"Your age is {age}.")
    print("You ara not born yet!")
elif age <= 15:
    print(f"Your age is {age}.")
    print("You are not eligible for driving license.")
    print("You are too young!")
else:
    print("Please enter a valid age number!")
```

```python
# Exercise 2: Positive or Negative Write a program that checks if a number is positive, negative, or zero.

num = int(input("Enter a number : "))

if num == 0:
    print(f"Your number is {num}.")
    print("It is neither positive nor negative.")
elif num >= 0:
    print(f"Your number is {num}.")
    print("It's a positive number!")
elif num <= 0:
    print(f"Your number is {num}.")
    print("It's a negative number!")
else:
    print("Please enter a valid number!")
```

```python
# Exercise 3: Larger Number Write a program that compares two numbers and prints which one is larger.

num1 = int(input("Enter first number :"))
num2 = int(input("Enter first number :"))

if num1 == num2:
    print("Both numbers are equal.")
elif num1 >= num2:
    print(f"{num1} is larger than {num2}.")
elif num2 >= num1:
    print(f"{num2} is larger than {num1}.")
else:
    print("Enter a valid number!")
```

```python
# Exercise 4: Calculator Discount A store offers:
# 10% discount if purchase > $100
# 5% discount if purchase > $50
# No discount otherwise
# Calculate the final price.

original_price = int(input("Enter original price : "))

if original_price >= 100:
    print(f"Original price : ${original_price}")
    print("Total discount = 10%")
    selling_price = original_price - (original_price / 100) * 10
    print(f"Final price : ${selling_price}")  
elif original_price >= 50:
    print(f"Original price : ${original_price}")
    print("Total discount = 5%")
    selling_price = original_price - (original_price / 100) * 5
    print(f"Final price : %{selling_price}")
elif original_price >= 0:
    print(f"Original price : ${original_price}")
    print(f"No discount for this price")
    print(f"Final price : ${original_price}")
else:
    print("Enter a valid amount")
    

```

```python
# Exercise 6: Grade Calculator with Bonus Calculate grades (A-F) but add +1 bonus grade level if attendance >
# 90%.

score = float(input("Enter your marks : "))
attendance = float(input("Enter your attandance percentage : "))
grades = ['F','D','C','B','A']

if score >= 90:
    grade_idx = 4
elif score >= 80:
    grade_idx = 3
elif score >= 70:
    grade_idx = 2
elif score >= 60:
    grade_idx = 1
else:
    grade_idx = 0

if attendance >= 90:
    if grade_idx <= 4:
        grade_idx += 1
print(grades[grade_idx])        
```

```python
# Exercise 7: Login System with Attempts Create a login system that:
# Checks username and password
# Tracks login attempts (max 3)
# Locks account after failed attempts

correct_username = "admin"
correct_password = "password123"

attempts = 0
max_attempts = 3
account_locked = False

print("---Secure Login System---")

while attempts < max_attempts:
    username = input("Enter username: ")
    password = input("Enter password: ")
    
    if username == correct_username and password == correct_password:
        print("\nLogin successful! Welcome back.")
        break
    else:
        attempts += 1
        remaining = max_attempts - attempts
        
        if remaining > 0:
            print(f"Invalid credentials. You have {remaining} attempts left.\n")
        else:
            account_locked = True
            print("\nToo many failed attenpts. Your accounts is now locked.")
            
if account_locked:
    print("Please contact support to reset your password.")

```

```python
# Exercise 8: Shipping Cost Calculator Calculate shipping based on:
# Weight of package
# Destination (domestic/international)
# Express or standard shipping

weight = float(input("Enter weght of the package (kg): "))
destination = input("Enter destination (domestic / international): ").strip().lower()
method = input("Enter shipping method (express / standard): ").strip().lower()

if destination == "domestic":
    cost = weight * 5.00
else:
    cost = weight * 15.00
    
if method == "express":
    cost += 10.00
    
print("_" * 40)
print(f"Summary: {destination.capitalize()} | {method.capitalize()}")
print(f"Total Shipping Cost: ${cost:.2f}")
print("-" * 40)
```
