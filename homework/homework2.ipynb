```python
# 1. Count how many vowels are in a given string.

word = "I'm learning python!"
vowels = "aeiou"
count = 0
for w in word:
    if w.lower() in vowels:
        count += 1
print(f"Vowels in string : {count}")
```

```python
# 2. Reverse a number using only a for loop (no slicing).
# Example: Input → 12345, Output → 54321

numbers = 12345
reverse = ""

for n in str(numbers):
    reverse = str(n) + str(reverse)
print(f"Reverse a number using only a for loop : {reverse}")
```

```python
# 3. Find the sum of digits of a number.
# Example: 5483 → 20

numbers = [5,4,8,3]
sum = 0

for num in numbers:
    sum += num
print(f"Sum of digits of numbers : {sum}")
```

```python
# 5. Find the factorial of a number.

number = int(input("Enter a number : "))
fact = 1

for i in range(1, number + 1):
    fact = fact * i
print(f"{number} factorial is {fact}")
    
```

```python
# 6. Check whether a number is prime using a for loop.

number = int(input("Enter a number : "))

if number >= 1:
    
    for i in range(2, number):
        if (number % i) == 0:
            print(f"{number} is not a prime number.")
            break
    else:
        print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
# 9. Find the largest number in a list.
# Example: [4, 7, 1, 9, 3]

number = [4, 7, 1, 9, 3]
max_number = number[0]

for num in number:
    if num > max_number:
        max_number = num
        
print(f"Largest number in a list : {max_number}")
    
```

```python
# 11. From 1 to 50, find:

# a) sum of even numbers

even_numbers = []
sum = 0
for i in range(0,51,2):
    even_numbers.append(i)
    sum += i
print(f"Sum of even numbers from 1 to 50 : {sum}")

# b) sum of odd numbers

odd_numbers = []
sum = 0
for i in range(1,51,2):
    odd_numbers.append(i)
    sum += i
print(f"Sum of odd numbers from 1 to 50 : {sum}")
```

```python
# 12. Remove all spaces from a string.
# Example: "data science rocks" → "datasciencerocks"

string = "data science rocks"
no_space = ""

for s in string:
    if s != " ":
        no_space += s
        
print(f"""String : {string}
Removing spaces : {no_space}""")
```

```python
# 13. Check whether a string is a palindrome.
# Example: "madam"

string = input("Enter a word : ")
is_palindrome = True

for i in range(len(string) // 2):
    if string.lower()[i] != string[-(i + 1)]:
        is_palindrome = False
        break
    
if is_palindrome:
    print("Its a Palindrome!")
else:
    print("Its not a Palindrome!")
```

```python
# 14. Find duplicate values in a list.
# Example: [1,2,3,2,4,5,1]

my_list = [1,2,3,2,4,5,1]
dup_value = []
seen = []

for value in my_list:
    if value in seen:
        if value not in dup_value:
            dup_value.append(value)
    else:
        seen.append(value)
        
print((dup_value))
```

```python
# 15. Print the frequency of each character in a string.
# Example: "banana"

string = input("Enter character : ")
frequency = {}

for i in string:
    if i in frequency:
        frequency[i] += 1
    else:
        frequency[i] = 1
print(frequency)
```
