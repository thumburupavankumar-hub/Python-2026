```python
# Question 11: Find the most frequent character in "mississippi"

word = "mississippi"
frequency = {}
for char in word:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
max_freq = max(frequency.values())
most_freq =[char for char, count in frequency.items() if count == max_freq]

print(f"Most frequent character : {most_freq}")
```

```python
# Question 39: Find the most common word in "the quick brown fox jumps over the lazy dog"

sen = "the quick brown fox jumps over the lazy dog"
words = sen.split()
counts = {}

for word in words:
    counts[word] = counts.get(word, 0) + 1
most_common = max(counts, key= counts.get)

print(f"Most common word : {most_common}")
```

```python
# Question 41: Check if string contains balanced parentheses: "((()))"

sen = "(((National Aeronautics and Space Administration)))"
balance = 0

for char in sen:
    if char == "(":
        balance += 1
    elif char == ")":
        balance -= 1
    if balance <= 0:
        break
    
if balance == 0:
    print("Balanced")
else:
    print("Unbalanced")
        
```

```python
# Question 42: Convert "hello world" to Morse code

morse_dict = {
    'a': '.-', 'b': '-...', 'c': '-.-.', 'd': '-..', 'e': '.', 'f': '..-.', 
    'g': '--.', 'h': '....', 'i': '..', 'j': '.---', 'k': '-.-', 'l': '.-..', 
    'm': '--', 'n': '-.', 'o': '---', 'p': '.--.', 'q': '--.-', 'r': '.-.', 
    's': '...', 't': '-', 'u': '..-', 'v': '...-', 'w': '.--', 'x': '-..-', 
    'y': '-.--', 'z': '--..', ' ': '/'
}

char = "hello world"
morse_code = ""

for i in char.lower():
    morse_code += morse_dict[i] + " "
print(morse_code)
```

```python
# Question 43: Find the longest common substring between "programming" and "grammar"

word1 = "programming"
word2 = "grammar"
longest = ""

for i in range(len(word2)):
    for j in range(i + 1, len(word2) + 1):
        subtracting = word2[i:j]
        if subtracting in word1 and len(subtracting) > len(longest):
            longest = subtracting
            
print(f"Longest common substring: {longest}")
```

```python
# Question 46: Convert "hello world" to Pig Latin

text = "hello world"
words = text.split()
pig_latin = []

for word in words:
    transform = word[1:] + word[0] + "ay"
    pig_latin.append(transform)
    
result = " ".join(pig_latin)
print(f"Pig Latin : {result}")
```

```python
# Question 47: Check if string is a valid IPv4 address: "192.168.1.1"

ipv4 = "192.168.1.1"
parts = ipv4.split(".")

if len(parts) == 4:
    is_valid = True
    for part in parts:
        if not (part.isdigit() and 0 <= int(part) <= 255):
            is_valid = False
            break
            
    if is_valid:
        print("It's valid")
    else:
        print("It's invalid (numbers out of range or not digits)")
else:
    print("It's invalid (must have 4 parts)")
```

```python
# Question 49: Convert "hello world" to ROT13 encoding

text = "hello world"
result = ""

for char in text:
    if 'a' <= char <= 'z':
        shifted = (ord(char) - ord('a') + 13) % 26
        result += chr(shifted + ord('a'))
    
    elif 'A' <= char <= 'Z':
        shifted = (ord(char) - ord('A') + 13) % 26
        result += chr(shifted + ord('A'))
    
    else:
        result += char

print(f"ROT13 encoding : {result}")
```

```python
# Question 50: Check if string is a valid credit card number: "4532015112830366"

c_num = "4532015112830366"

digits = []
for d in c_num:
    digits.append(int(d))

total_sum = 0
reverse_digits = digits[::-1]

for i in range(len(reverse_digits)):
    n = reverse_digits[i]
    
    if i % 2 == 1:
        n = n * 2

        if n > 9:
            n -= 9
            
    total_sum += n

if total_sum % 10 == 0:
    print(f"{c_num} is a VALID credit card number.")
else:
    print(f"{c_num} is INVALID.")
```
