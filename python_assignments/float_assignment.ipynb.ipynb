```python
# Question 1: Calculate the average of 3.14, 2.718, 1.618, 0.577

num = [3.14, 2.718, 1.618, 0.577]
average = sum(num) / len(num)
print(f"Average: {average}")
```

```python
# Question 1: Calculate the average of 3.14, 2.718, 1.618, 0.577

num = [3.14, 2.718, 1.618, 0.577]
l = len(num)
sum = 0
for i in num:
    sum += i
avg = sum / l
print(f"Average = {avg}")
```

```python
# Question 2: Convert 98.6 Fahrenheit to Celsius (F = C * 9/5 + 32)

F = float(input("Enter Fahrenheit: "))
C = (F-32) * 5/9
print(f"Temp of celcius : {C:.2f} degrees celcius")
```

```python
# Question 3: Calculate the compound interest on $1000 at 5.5% for 3 years
# Amount= p*(1+r/100)**time
#CI=amount- principle
principal = 1000
rate = 5.5
time = 3
amount = principal*(1 + rate / 100)** time
compound_interest = amount - principal
print(f"compound_interest : ${compound_interest:.2f}")
```

```python
# Question 4: Find the hypotenuse of a right triangle with sides 3.5 and 4.2

import math

l = 3.5
b = 4.2

hypotenuse = math.sqrt(l**2 + b**2)
print(f"Hypotenus : {hypotenuse:.2f}")
```

```python
# Question 5: Calculate the volume of a sphere with radius 7.8

import math

r = 7.8
v = (4/3) * math.pi * r ** 3
print(f"Volume if sphere : {v:.2f}")
```

```python
# Question 9: Calculate the simple interest: Principal=2500, Rate=6.5%, Time=2.5 years

principal = 2500
rate = 6.5
time = 2.5
SI = (principal * rate * time) / 100
print(f"Simple Interest : {SI:.2f}")
```
